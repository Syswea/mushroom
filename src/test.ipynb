```python

# ========================
# 2. è°ƒç”¨API
# ========================
def call_api(data: List[Dict[str, Any]]) -> List[Dict[str, Any]]:
    """å‘é€è¯·æ±‚åˆ°APIå¹¶è¿”å›ç»“æœ"""
    url = "http://localhost:8000/predict"
    headers = {"Content-Type": "application/json"}
    
    try:
        response = requests.post(url, json=data, headers=headers)  # ä½¿ç”¨jsonå‚æ•°è‡ªåŠ¨åºåˆ—åŒ–
        response.raise_for_status()
        return response.json()
    except requests.exceptions.RequestException as e:
        print(f"âŒ APIè¯·æ±‚å¤±è´¥: {e}")
        print(f"é”™è¯¯è¯¦æƒ…: {e.response.text if e.response else 'No response'}")
        raise
```

```python

# ========================
# 3. å±•ç¤ºç»“æœ
# ========================
def display_result(result: Dict[str, Any]):
    """æ ¼å¼åŒ–æ‰“å°å•ä¸ªé¢„æµ‹ç»“æœ"""
    print("\n" + "="*50)
    print("ğŸ”¥ æ¯’è˜‘è‡é¢„æµ‹ç»“æœ (æ¦‚ç‡ > 0.5 = æ¯’æ€§)")
    print("="*50)
    
    # æå–å…³é”®ä¿¡æ¯
    predicted_class = result["predicted_class"]
    probability = result["probability_poisonous"]
    
    # ç”Ÿæˆåˆ†ç±»æ ‡ç­¾
    class_label = "æ¯’è˜‘è‡ (p)" if predicted_class == "p" else "å¯é£Ÿç”¨ (e)"
    toxicity = "æ˜¯" if probability > 0.5 else "å¦"
    
    # æ ¼å¼åŒ–è¾“å‡º
    print(f"è˜‘è‡ID: {result['id']}")
    print(f"é¢„æµ‹ç±»åˆ«: {class_label}")
    print(f"æ¯’æ€§æ¦‚ç‡: {probability:.2%} ({probability:.4f})")
    print(f"æ˜¯å¦æ¯’æ€§: {toxicity}")
    print("="*50)
```
